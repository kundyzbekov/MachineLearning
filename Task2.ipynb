```python
from scipy.linalg import solve
from matplotlib import pylab as plt
import math
%matplotlib inline
def func(x):
    return math.sin(x/5.0) * math.exp(x/10.0) + 5 * math.exp(-x/2.0)
```

```python
import numpy as np
f1 = func(1)
f15 = func(15)
print (f1, ' ', f15)
```

```python
b = np.array([f1, f15])
a = np.array([[1, 1], [1, 15]])
w = solve(a, b)
print(w)
```

```python
x = np.arange(1, 16)
```

```python
yp = w[0] + w[1] * x
yf = [func(f) for f in x]
```

```python
plt.plot(x, yp)
plt.plot(x, yf)
plt.show()
```

```python
b = np.array([func(1), func(8), func(15)])
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
w = solve(a, b)
print(w)
```

```python
yp = w[0] + w[1] * x + w[2] * (x ** 2)
```

```python
plt.plot(x, yp)
plt.plot(x, yf)
plt.show()

```

```python
b = np.array([func(1), func(4), func(10), func(15)])
a = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 225*15]])

```

```python
w = solve(a, b)
print(w)

```

```python
yp = w[0] + w[1] * x + w[2] * (x ** 2) + w[3] * (x ** 3)
```

```python
plt.plot(x, yp)
plt.plot(x, yf)
plt.show()
```
